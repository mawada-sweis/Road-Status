```python
data = pd.read_csv('../data/raw/data.csv')
```

```python
# Remove unnecessary columns based on usefulness of it
FS.delete_unnecessery_columns(data=data)
# Display the columns after cleaned
data.columns
```

```python
# Extract the date and time to separate columns
data['year'], data['month'], data['day'], data['hour'], data['minute'], data['second'] = FS.extract_dates(date=data['date'])
# Remove the date column cause it has become useless
data.drop(columns='date', inplace=True)
# Display the columns after cleaned
data
```

```python
# Extract the ID of the user who sent the message
data['from_id'] = data['from_id'].apply(lambda x: FS.extract_id(json_string=x, target_id='user_id'))
# Display the columns after cleaned
data
```

```python
# Extract the ID of the message being replied to, if any.
data['reply_to'] = data['reply_to'].apply(lambda x: FS.extract_id(json_string=x, target_id='reply_to_msg_id'))
# Display the columns after cleaned
data
```
